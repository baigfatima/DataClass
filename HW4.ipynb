```python
import pandas as od
import numpy as np

source = "Downloads/purchase_data.csv"

purchase_data = pd.read_csv(source)
```

```python
#Player Count
player_count = purchase_data["SN"].nunique()

total_player_df = pd.DataFrame(data={"Total Players":[player_count]})
total_player_df
```

```python
#Purchasing Analysis (Total)
uniq_purch = purchase_data["Item ID"].nunique()

avg_price = round(purchase_data["Price"].mean(),2)

total_purch = purchase_data["Purchase ID"].nunique()

total_rev = purchase_data["Price"].sum()

purch_analysis_total = pd.DataFrame(data={
                                        "Number of Unique Items":[uniq_purch],
                                        "Average Price":[avg_price], 
                                        "Number of Purchases":[total_purch], 
                                        "Total Revenue":[total_rev]
                                        })

purch_analysis_total.style.format({'Average Price': '${:.2f}', 'Total Revenue': '${:.2f}'})
```

```python
#Gender Demographics
df = purchase_data.drop_duplicates(["SN"])

gender_total = df["Gender"].value_counts()

gender_player_count = round(gender_total/player_count*100,2)

gender_df = pd.DataFrame({"Total":gender_total, "Percentage of Players":gender_player_count})
gender_df
```

```python
#Purchasing Analysis (Gender)
df = purchase_data.groupby("Gender")

purch_count = df["SN"].count()

avg_price = round(df["Price"].mean(),2)

total_price = df["Price"].sum()

avg_total_purch = round(total_price/gender_total, 2)

gender_analysis_df = pd.DataFrame({
                                "Purchase Count":purch_count, 
                                "Average Purchase Price":avg_price, 
                                "Total Purchase Value":total_price, 
                                "Average Total Purchase Per Person": avg_total_purch
                                })

gender_analysis_df.style.format({'Average Purchase Price': '${:.2f}', 'Total Purchase Value': '${:.2f}', "Average Total Purchase Per Person": '${:.2f}'})
```

```python
#Age Demographics
bins = [0, 10, 15, 20, 25, 30, 35, 40, 45]

groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
          
age_demo = purchase_data.drop_duplicates(["SN"])
          
age_demo["Age Group"] = pd.cut(age_demo["Age"], bins, labels=groups)
          
age_demo = age_demo.groupby("Age Group").count().drop(axis=1,columns=["Purchase ID", "SN", "Gender", "Item ID", "Item Name", "Price"]).rename(index=str, columns={"Age":"Total Count"})
          
age_demo["Percentage of Players"]= round(age_demo["Total Count"]/player_count*100,2) 
del age_demo.index.name
age_demo
```

```python
#Purchasing Analysis (Age) 
df = purchase_data

df["Age Group"] = pd.cut(df["Age"], bins, labels=groups)

df = df.groupby("Age Group")

purch_count = df["SN"].count()

avg_price = round(df["Price"].mean(),2)

total_price = df["Price"].sum()

avg_total_purch = round(total_price/age_demo["Total Count"],2)

age_analysis_df = pd.DataFrame({
                        "Purchase Count":purch_count,
                        "Average Purchase Price":avg_price, 
                        "Total Purchase Value":total_price, 
                        "Average Total Purchase Per Person": avg_total_purch
                        })
del age_analysis_df.index.name

age_analysis_df.style.format({'Average Purchase Price': '${:.2f}', 'Total Purchase Value' : '${:.2f}',"Average Total Purchase Per Person": '${:.2f}'})
```

```python
#Top Spender
df = purchase_data.groupby("SN")

purch_count =df["SN"].count()

avg_price = round(df["Price"].mean(),2)

total_price = df["Price"].sum()

age_analysis_df = pd.DataFrame({
                        "Purchase Count":purch_count,
                        "Average Purchase Price":avg_price, 
                        "Total Purchase Value":total_price
                        })
del age_analysis_df.index.name

age_analysis_df.style.format({'Average Purchase Price': '${:.2f}', 'Total Purchase Value' : '${:.2f}'})

age_analysis_df.sort_values(by=["Total Purchase Value"],axis=0, ascending=False).head(15)
```

```python
#Most Popular Items
df = purchase_data[["Item ID", "Item Name", "Price"]]

#dataframe grouped by item id and item name
most_popular_group = df.groupby(["Item ID", "Item Name"])

#total purchase count of items
purch_count = most_popular_group["Price"].count()

#total purchase value of an item
total_price = most_popular_group["Price"].sum()

#price of items
price = total_price/purch_count

#create most popular irem data frame
most_popular_item = pd.DataFrame({
                        "Purchase Count":purch_count,
                        "Item Price":price,
                        "Total Purchase Value":total_price
                        })
#declare most profitable data frame ine new variable
most_profitable_item = most_popular_item


mpopi_df = most_popular_item.sort_values(by=["Purchase Count"],axis=0, ascending=False)

#format dollar for data frame
mpopi_df.style.format({'Item Price': '${:.2f}', 'Total Purchase Value' : '${:.2f}'})
mpopi_df.head(5)
```

```python
#Most Profitable Items
mprofi_df = most_profitable_item.sort_values(by=["Total Purchase Value"],axis=0, ascending=False)

mprofi_df.style.format({'Item Price': '${:.2f}', 'Total Purchase Value' : '${:.2f}'})
mprofi_df.head(5)
```
